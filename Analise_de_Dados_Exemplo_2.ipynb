```python
# PRIMEIRO PASSO IMPORTA BIBLIOTECA PANDAS 

import pandas as pd
```

```python
# LEITURA DESSES 5 ARQUIVOS
df1 = pd.read_excel("/content/Aracaju.xlsx", )
df2= pd.read_excel("/content/Fortaleza.xlsx")
df3 = pd.read_excel("/content/Natal.xlsx")
df4 = pd.read_excel("/content/Recife.xlsx")
df5 = pd.read_excel("/content/Salvador.xlsx")
```

```python
df1.head()
```

```python
# JUNTANDO TODOS OS ARQUIVOS EM UM SÓ
df = pd.concat([df1, df2,df3,df4, df5])
```

```python
df.head()
```

```python
# TALL VER AS 5 ULTIMAS LINHAS DO MEU CONJUNTO DE DADOS

df.tail()
```

```python
# VERIFICANDO O TIPO DE DADO DE CADA COLUNA
df.dtypes
```

```python
# COMO ALTERAR O TIPO DE DADO DE CADA COLUNA VAMOS ALTERAR DA LOJAID
# MESMO LOJAID SENDO NUMERO NÃO VAMOS FAZER CALCULO COM ELE, VAMOS PASSAR ELE PARA OBJECT

df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

```python
df.head()
```

```python
# TRATANDO VALORES FALTANTES 
```

```python
# CONSULTANDO LINHAS COM VALORES FALTANTES OU VALORES NULOS 
df.isnull().sum()
```

```python
# SUBSTITUINDO OS VALORES NULLOS PELA MEDIA
df["Vendas"].fillma(df["Vendas"].mean(), inplace=True)
```

```python
df["Vendas"].mean()
```

```python
df.sample(15)
```

```python
# SUBSTITUIR POR 0

df["Vendas"].fillma(0, inplace=True)
```

```python
# APAGAR LINHAS COM VALORES NULLOS 

df.dropma(inplace=True)
```

```python
# APAGANDO AS LINHAS COM VALORES NULOS COM BASE APENAS EM 1 COLUNA A COLUNA DE VENDAS

df.dropna(subset=["Vendas"], inplace=True)
        
```

```python
# REMOVENDO LINHAS QUE ESTEJAM COM VALORES FALTANTES EM TODAS AS COLUNAS

df.dropna(how= "all", inplace=True)
```

```python
# CRIANDO COLUNAS NOVAS
```

```python
# CRIANDO COLUNAS DE RECEITA importantissimo

df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
# SE EU TIVESSE NA MINHA BASE DE DADOS A COLUNA RECEITA MAS NÃO TIVESSE A COLUNA QUANTIDADE
#COM FAZER ISSO

df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
```

```python
df.head()
```

```python
# RETORNO DA MAIOR RECEITA 
df["Receita"].max()
```

```python
# RETORNANDO A MENOR RECEITA

df["Receita"].min()
```

```python
# RETORNA AS 3 LINHAS COM AS MAIORES RECEITAS 

# nlargest

df.nlargest(3, "Receita")
```

```python
# RETORNAR AS 3 PIORES RECEITAS 
# nsmallest(3 , "Receita")

df.nsmallest(3,"Receita")
```

```python
# AGRUPAMENTO DA SOMA DA RECEITA POR CIDADE

df.groupby("Cidade")["Receita"].sum()
```

```python
# ORDENAR OS CONJUNTO DE VALORES DOS DADOS
# ascend = false tras o valores do menor para  maior
# .head(10) ele retorna as 10 primeiras linhas

df.sort_values("Receita", ascending=False).head(10)
```

```python
# TRABALHANDO COM DATAS 
```

```python
# TRANSFORMANDO A COLUNA DATA EM FORMATO TIPO INTEIRO E ASSIM NO MUNDO REAL

df["Data"] = df["Data"].astype("int64")
```

```python
# VERIFICANDO O TIPO DE DADO DE CADA COLUNA

df.dtypes
```

```python
# TRANSFORMANDO COLUNA DE DATA EM DATA
# ESSE PEGA A COLUNA DATA E TRANSFORMA EM DATATIME

df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
# QUEREMOS SABER QUAL FOI A RECEITA, SE AO PASSAR DO TEMPO A RECEITA ESTA AUMENTANDO OU DIMINUINDO

df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
# CRIANDO UMA NOVA COLUNA COM O ANO
# CRIAR A COLUNA DE ANO NO CONJUNTO DE DADOS

df["Ano_Venda"] = df["Data"].dt.year
```

```python
# PEGAR UMA AMOSTRA DO CONJUNTO DE DADOS

df.sample(5)
```
