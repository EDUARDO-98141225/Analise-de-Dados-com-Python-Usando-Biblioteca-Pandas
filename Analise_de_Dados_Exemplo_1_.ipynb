```python
# PROXIMO METODO COMO VER OS CONTINENTES DO MEU CONJUNTO DE DADOS
df["Continente"].unique()
```

```python
# RETORNAR APENAS OS VALORES DO CONTINENTE OCEANIA COMO FAZER ABAIXO

Oceania = df.loc[df["Continente"] == "Oceania"]
Oceania.head()
```

```python
# LOC REALIZA FILTRO NA NOSSA BASE DE DADOS 
Oceania["Continente"].unique()
```

```python
# AGRUPAMENTO DE DADOS USAR O GROUPBY
# QUANTOS PAÍSES NOS TEMOS PARA CADA CONTINENTE USAR O NUNIQUE, ELE FAZ UMA CONTAGEM DISTINTA

df.groupby("Continente")["País"].nunique()
```

```python
# PARA CADA ANO QUAL É A   ESPECTATIVA DE VIDA MEDIA DE CADA PAÍS NO NOSSO CONJUNTO DE DADOS 
df.groupby("Ano")["Expectativa de vida"].mean()
```

```python
# OUTRA FORMA DE FAZER ESTES CALCULOS
# MEDIA DO PIB
df["PIB"].mean()
```
